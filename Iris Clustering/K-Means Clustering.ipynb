# K-Means Clustering
---
### Using the *iris* dataset

```python
#importing all the libraries
from sklearn.cluster import KMeans
from sklearn import datasets
from sklearn.metrics import accuracy_score,confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
#mapping of indices and categories
imap={'setosa':0, 'versicolor':1, 'virginica':2}
i_imap={'0':'setosa', '1':'versicolor','2':'virginica'}
```

```python
#load the iris dataset
iris_df=sns.load_dataset('iris')
```

```python
#plot relations between the attributes

sns.pairplot(iris_df, hue="species" , palette="husl")
```

```python
x=iris_df.values[:,:4].astype(float)
y=np.array([imap.get(str(x)) for x in iris_df.values[:,4]]).astype(int)
```

```python
est=KMeans(n_clusters=3)
```

```python
est.fit(x)
```

```python
y_p=est.predict(x)
```

```python
new_iris=iris_df.copy()
new_iris['species']=np.array([i_imap.get(str(i)) for i in y_p])
```

```python
sns.pairplot(new_iris, hue="species", palette="husl" )
```

```python
print("Confusion Matrix - \n",confusion_matrix(y, y_p))
print("\nAccuracy - %.5f" %(accuracy_score(y, y_p)))
```
