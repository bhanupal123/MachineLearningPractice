# Multivariate Linear Regression using scikit-learn
---

```python
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
#creating an instance of the dataset
boston=datasets.load_boston()

#description of the dataset
print("Shape of the features - ")
print(boston.data.shape)

print("Shape of the targets - ")
print(boston.target.shape)
```

## We remove feature 4 and 9 as they seem to be categorical in nature

```python
np.delete(boston.data,3,1)
```

```python
np.delete(boston.data,8,1)
```

```python
#dividing the dataset into training and testing
boston_train=boston.data[:-20]
boston_test=boston.data[-20:]
```

```python
#dividing the dataset into training and testing
boston_train_y=boston.target[:-20]
boston_test_y=boston.target[-20:]
```

```python
#create an object of Linear Regression
regr = linear_model.LinearRegression()
```

```python
#Train the model using the training sets
regr.fit(boston_train, boston_train_y)
```

```python
#The coefficients
print('Coefficients: \n', regr.coef_)

#The mean square error
print("\nResidual sum of squares: %.2f"
      % np.mean((regr.predict(boston_test) - boston_test_y) ** 2))
```

```python
#Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(boston_test, boston_test_y))
```

```python
#Plot outputs
plt.figure(2)
plt.figure(figsize=(12,28))

plt.subplot(621)
plt.scatter(boston_test[:,0],boston_test_y, color='red')
plt.plot(boston_test[:,0], regr.predict(boston_test), linewidth=1)
plt.title("Feature 1")

plt.subplot(622)
plt.scatter(boston_test[:,1],boston_test_y, color='red')
plt.plot(boston_test[:,1], regr.predict(boston_test), linewidth=1)
plt.title("Feature 2")

plt.subplot(623)
plt.scatter(boston_test[:,2],boston_test_y, color='red')
plt.plot(boston_test[:,2], regr.predict(boston_test), linewidth=1)
plt.title("Feature 3")

plt.subplot(624)
plt.scatter(boston_test[:,3],boston_test_y, color='red')
plt.plot(boston_test[:,3], regr.predict(boston_test), linewidth=1)
plt.title("Feature 4")

plt.subplot(625)
plt.scatter(boston_test[:,4],boston_test_y, color='red')
plt.plot(boston_test[:,4], regr.predict(boston_test), linewidth=1)
plt.title("Feature 5")

plt.subplot(626)
plt.scatter(boston_test[:,5],boston_test_y, color='red')
plt.plot(boston_test[:,5], regr.predict(boston_test), linewidth=1)
plt.title("Feature 6")

plt.subplot(627)
plt.scatter(boston_test[:,6],boston_test_y, color='red')
plt.plot(boston_test[:,6], regr.predict(boston_test), linewidth=1)
plt.title("Feature 7")

plt.subplot(628)
plt.scatter(boston_test[:,7],boston_test_y, color='red')
plt.plot(boston_test[:,7], regr.predict(boston_test), linewidth=1)
plt.title("Feature 8")

plt.subplot(629)
plt.scatter(boston_test[:,8],boston_test_y, color='red')
plt.plot(boston_test[:,8], regr.predict(boston_test), linewidth=1)
plt.title("Feature 9")

plt.subplot(6,2,10)
plt.scatter(boston_test[:,9],boston_test_y, color='red')
plt.plot(boston_test[:,9], regr.predict(boston_test), linewidth=1)
plt.title("Feature 10")

plt.subplot(6,2,11)
plt.scatter(boston_test[:,10],boston_test_y, color='red')
plt.plot(boston_test[:,10], regr.predict(boston_test), linewidth=1)
plt.title("Feature 11")

plt.show()
```
